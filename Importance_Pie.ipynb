```python
import pandas as pd
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')
```

```python
# Specify the path to the Excel file in your Google Drive
file_path = '/content/drive/MyDrive/Bootcamp/Project4/Data/Features.xlsx'

# Specify the name or index of the sheet you want to read
sheet_name = 'Actual'

# Read the Excel file
features_df = pd.read_excel(file_path, sheet_name=sheet_name, header=None)

# Add your own column names (header)
column_names = ["Name", "Importance Per ML"]
features_df.columns = column_names

# Remove "Importance of feature from rows"
features_df["Name"] = features_df["Name"].str.replace('Importance of feature ', '')

# Display the first few rows of the DataFrame
features_df.head()
```

```python
# Set a threshold for importance values
importance_threshold = 0.01
```

```python
filtered_df = features_df[features_df['Importance Per ML'] > importance_threshold]
```

```python
import matplotlib.pyplot as plt

# create labels and values from the dataframe
labels = filtered_df['Name']
values = filtered_df['Importance Per ML']

# create pie chart
plt.figure(figsize=(8, 8))
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Feature Importances')
plt.show()
```
